```python
#Remove links
```

```python
df = df['original_text'] = df.original_text.replace(
            r'http\S+', '', regex=True).replace(r'www\S+', '', regex=True)
```

```python
df = df['original_text'] = df['original_text'].str.replace(
            "[^a-zA-Z#]", " ")
```
